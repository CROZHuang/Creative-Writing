```python
# MIND THE FILE NAME
file_path = '/content/drive/MyDrive/Creativity/Qwen2.5_7b_16_FT_WritingPrompts_full.xlsx'
df = pd.read_excel(file_path, sheet_name ='Sheet1', engine = 'openpyxl')
```

```python
# input data
fluency_data = df['Fluency']
flexibility_data = df['Flexibility']
originality_data = df['Originality']
elaboration_data = df['Elaboration']
TTCT_add = df['TTCT']
TTCT_multi = df['TTCT_M']

# Compute the mean and variance
mean_fluency = fluency_data.mean()
variance_fluency = fluency_data.std()
mean_flexibility = flexibility_data.mean()
variance_flexibility = flexibility_data.std()
mean_originality = originality_data.mean()
variance_originality = originality_data.std()
mean_elaboration = elaboration_data.mean()
variance_elaboration = elaboration_data.std()
mean_TTCT = TTCT_add.mean()
variance_TTCT = TTCT_add.std()
mean_TTCT_M = TTCT_multi.mean()
variance_TTCT_M = TTCT_multi.std()

# Display the results
print('name of the model:Qwen2.5_7b_16_WritingPrompts')
print('------')
print(f"Mean of Fluency: {mean_fluency}")
print(f"SD of Fluency: {variance_fluency}")
print('------')
print(f"Mean of Flexibility: {mean_flexibility}")
print(f"SD of Flexibility: {variance_flexibility}")
print('------')
print(f"Mean of Originality: {mean_originality}")
print(f"SD of Originality: {variance_originality}")
print('------')
print(f"Mean of Elaboration: {mean_elaboration}")
print(f"SD of Elaboration: {variance_elaboration}")
print('------')
print(f"Mean of TTCT_Score(algorithmic): {mean_TTCT}")
print(f"SD of TTCT_Score(algorithmic): {variance_TTCT}")
print('------')
print(f"Mean of TTCT_Score(geometric): {mean_TTCT_M}")
print(f"SD of TTCT_Score(geometric): {variance_TTCT_M}")
```

```python
# Plot the box plot for fluency data
plt.figure(figsize=(10, 5))
plt.boxplot([fluency_data.dropna(), flexibility_data.dropna(), originality_data.dropna(), elaboration_data.dropna(), TTCT_add.dropna(), TTCT_multi.dropna()],
            vert=True,
            labels=['Fluency', 'Flexibility', 'Originality', 'Elaboration', 'TTCT_Score(+)', 'TTCT_Score(*)'])
plt.title('Box Plot of TTCT for Qwen2.5_7b_16_WritingPrompts')
plt.xlabel('Creativity Metrics')
plt.ylim(-0.05,1.05)
plt.grid(True)
plt.show()
```
